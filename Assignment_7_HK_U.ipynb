Nearby points get high probabilities and far away points get low probabilities.

t-SNE is an imporved version of stochastic neighbour embedding

Basic idea of SNE is that The points that are close to each other in high dimension space should also be close to each other in lower dimensions.

The measured distances should for hpently be very simular in lower dimension
the loss function $Loss(HighDim, LowDim)$ should tell us differences in them are

We want to minimize this to improve our low-dimensional representation:

$$
\frac{ \partial (HighDim, LowDim) }{ \partial LowDim}
$$

The first step in the SNE algorithm is to turn the distences between a point and its neighbours into a probability distribution. (Gaussian distribution in this instantce)
We convert distences into gaussian distribution.
We can think of this as the probability of a blue point being close to (neighbor of) the red point (center point)



Use this to describe the probability:
$$
p_{0,j} = e^{-\frac{||x_0-x_j||^2}{2 \sigma^2_0}}
$$
Here the shape $\sigma$ defines to a big degree the shape of the probability curve.

We adjust this curve based on the hyperparameter called perplexity


Perplexity in Stochastic Neighbor Embedding (SNE) and t-SNE is a crucial hyperparameter that defines the expected density of the data, acting as a smooth measure of the number of effective nearest neighbors each point considers when mapping high-dimensional data to a lower-dimensional space.


To make all the probabilities add up to 1 wo normalize each gaussian as the sum of all the others(this can be a very costly operation):


$$
 p_{j|i} = \frac{(1+d^2_{ij})^{-1}}{\sum_{kl}(1+d^2_{ij})^{-1}}
$$


The most common way to measure how simular two probability distributions are is to use the Kullback-Leibler divergence:


$$
D_{KL}(P||Q) = \sum_i P(i) log \frac{P(i)}{Q(i)}
$$
If this is large -> distributions are very different.


If it is small -> distributions are very similar.

If two distributions $Q$ and $P$ matches perfectly. Kullback-Leibler is 0


the gradient $ \frac{\partial D_{KL}(P||Q)}{\partial y_i}$ with respect to our low dimensional embeddings allows us to adjust the low-dimensional representation to be closer to our original distribution.

UMAP is a much newer technique for dimensionality rediction because: "UMAP is generally considered superior to t-SNE for visualization because it is significantly faster, scales better to large datasets, and preserves more of the global data structure. While PCA is much faster for linear, large-scale structures, UMAP excels at capturing non-linear relationships and local clusters better"




It doesn't have a formula to "solve" for the best coordinates. Instead, it starts with a random guess of where the points should be in the low-dimensional map. Then, it slowly moves the points around, step-by-step (iteration-by-iteration), to make the low-dimensional similarities ($q_{ij}$) match the high-dimensional ones ($p_{ij}$). This is like trying to find the lowest point in a dark, hilly landscape by taking small steps and feeling the ground beneath you.


At its core, t-SNE needs to calculate the similarity between every pair of points. If you have 10,000 data points, that's about 50 million pairs. It has to do this repeatedly. This quadratic complexity ($N^2$) is the main bottleneck.


The statement about the t-distribution being easier because it "does not include the exponential function" is a bit of a side detail.
````python
# ===========================================================
# SETUP -- Run this cell first!
# ===========================================================
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots

from sklearn.manifold import TSNE, MDS, trustworthiness
from sklearn.decomposition import PCA
from sklearn.datasets import (make_swiss_roll, make_blobs, make_moons,
                               make_s_curve, load_digits)
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import pdist, squareform

from ipywidgets import interact, widgets
from IPython.display import display, HTML
from time import perf_counter

# -- Consistent color scheme --
COLORS = {"A": "#E74C3C", "B": "#2E86C1", "C": "#27AE60",
          "D": "#F39C12", "ntnu": "#00509E", "gray": "#AAAAAA"}
PLOTLY_DISCRETE = px.colors.qualitative.G10   # safe 10-color palette
RNG = np.random.RandomState(42)

# -- Reusable helper: generate datasets --
def make_dataset(name="swiss_roll", n=500, noise=0.5, seed=42):
    """Generate a synthetic dataset.  Returns (X, color, title)."""
    rng = np.random.RandomState(seed)
    if name == "swiss_roll":
        X, c = make_swiss_roll(n_samples=n, noise=noise, random_state=seed)
        return X, c, "Swiss Roll (3D)"
    elif name == "blobs":
        X, c = make_blobs(n_samples=n, centers=4, n_features=10,
                          cluster_std=1.5, random_state=seed)
        return X, c, "4 Gaussian Blobs (10D)"
    elif name == "moons":
        X, c = make_moons(n_samples=n, noise=0.08, random_state=seed)
        return np.hstack([X, rng.randn(n, 8) * 0.3]), c, "Two Moons (10D)"
    elif name == "s_curve":
        X, c = make_s_curve(n_samples=n, noise=noise, random_state=seed)
        return X, c, "S-Curve (3D)"
    elif name == "concentric":
        theta = rng.uniform(0, 2*np.pi, n)
        phi   = rng.uniform(0, np.pi, n)
        label = (np.arange(n) > n//2).astype(int)
        r = np.where(label == 0, 1.0, 3.0) + rng.randn(n)*0.15
        X = np.column_stack([r*np.sin(phi)*np.cos(theta),
                             r*np.sin(phi)*np.sin(theta),
                             r*np.cos(phi)])
        return X, label, "Concentric Spheres (3D)"

# -- Reusable helper: run t-SNE --
def run_tsne(X, perplexity=30, lr=200, n_iter=1000,
             init="pca", seed=42, early_exag=12.0):
    """Run t-SNE and return 2-D embedding (scikit-learn >= 1.6)."""
    tsne = TSNE(n_components=2,
                perplexity=min(perplexity, len(X)-1),
                learning_rate=lr if lr != "auto" else "auto",
                max_iter=n_iter, init=init,
                random_state=seed, early_exaggeration=early_exag)
    return tsne.fit_transform(X)

# -- Reusable Plotly scatter --
def plotly_scatter(Y, color, title="", continuous=False, size=5, opacity=0.75):
    """Quick Plotly 2-D scatter with color."""
    if continuous:
        fig = px.scatter(x=Y[:,0], y=Y[:,1], color=color,
                         color_continuous_scale="Spectral", title=title,
                         opacity=opacity)
    else:
        fig = px.scatter(x=Y[:,0], y=Y[:,1],
                         color=[str(c) for c in color],
                         title=title, opacity=opacity,
                         color_discrete_sequence=PLOTLY_DISCRETE)
    fig.update_layout(xaxis_title="Dim 1", yaxis_title="Dim 2",
                      width=700, height=500, template="plotly_white")
    fig.update_traces(marker=dict(size=size))
    return fig

print("Setup complete -- all packages loaded.")
````

## 1. Motivation: Why Do We Need Nonlinear Embeddings?

**What we already know from earlier lectures:**

| Method | Type | Preserves |
|--------|------|----------|
| MLR | Supervised, linear | Prediction accuracy |
| PCA / PCR | Unsupervised, linear | Global variance |
| ICA | Unsupervised, linear | Statistical independence |
| PLSR | Supervised, linear | Covariance with response |

**Today's challenge:** Real datasets often live on curved manifolds (spirals, S-curves).
Linear methods squash and mix local neighborhoods when projecting.

> **Goal:** Find a *nonlinear* map that "unrolls" the data and preserves
> **who is near whom** -- local neighborhood structure.

### When to use t-SNE and when NOT to

| Use for | Not for |
|-----------|----------|
| Exploring clusters | Measuring distances between clusters |
| Spotting outliers | Comparing cluster sizes |
| Inspecting learned representations | Quantitative global geometry |

````python
# -- 1a. PCA vs t-SNE on different datasets --
# PCA preserves global variance but squashes nonlinear structure.
# t-SNE preserves local neighborhoods even on curved manifolds.

def demo_motivation(dataset="swiss_roll", n_points=500):
    X, color, title = make_dataset(dataset, n=n_points)
    X_s = StandardScaler().fit_transform(X)
    is_cont = not np.array_equal(color, color.astype(int))

    # PCA embedding
    Y_pca = PCA(n_components=2).fit_transform(X_s)
    # t-SNE embedding
    Y_tsne = run_tsne(X_s, perplexity=30, seed=42)

    fig = make_subplots(rows=1, cols=2,
                        subplot_titles=["PCA (linear)", "t-SNE (nonlinear)"])
    if is_cont:
        fig.add_trace(go.Scatter(x=Y_pca[:,0], y=Y_pca[:,1], mode="markers",
                                 marker=dict(size=4, color=color, colorscale="Spectral",
                                             opacity=0.7)), row=1, col=1)
        fig.add_trace(go.Scatter(x=Y_tsne[:,0], y=Y_tsne[:,1], mode="markers",
                                 marker=dict(size=4, color=color, colorscale="Spectral",
                                             opacity=0.7)), row=1, col=2)
    else:
        for c_val in np.unique(color):
            mask = color == c_val
            clr = PLOTLY_DISCRETE[int(c_val) % len(PLOTLY_DISCRETE)]
            fig.add_trace(go.Scatter(x=Y_pca[mask,0], y=Y_pca[mask,1], mode="markers",
                                     marker=dict(size=4, color=clr, opacity=0.7),
                                     name=f"Class {int(c_val)}", showlegend=True),
                          row=1, col=1)
            fig.add_trace(go.Scatter(x=Y_tsne[mask,0], y=Y_tsne[mask,1], mode="markers",
                                     marker=dict(size=4, color=clr, opacity=0.7),
                                     name=f"Class {int(c_val)}", showlegend=False),
                          row=1, col=2)

    fig.update_layout(title_text=f"Motivation: {title}", width=950, height=450,
                      template="plotly_white")
    fig.show()
    print(f"Notice how PCA mixes structure on '{title}' while t-SNE separates neighborhoods.")

interact(demo_motivation,
         dataset=widgets.Dropdown(
             options=["swiss_roll", "s_curve", "blobs", "moons", "concentric"],
             value="swiss_roll", description="Dataset:"),
         n_points=widgets.IntSlider(min=200, max=800, step=100, value=500,
                                    description="N:"))
````

## 2. Key Terminologies

### 2a. Neighborhood
The subset of points that are "locally" similar to a given point $x_i$.
Linear methods (PCA, ICA) prioritize global variance/independence.
In complex data, **local topology** (who is next to whom) is often more informative.

### 2b. Embedding
Given $\{x_i\}_{i=1}^n$ with $x_i \in \mathbb{R}^D$, find $\{y_i\}_{i=1}^n$ with $y_i \in \mathbb{R}^d$ (typically $d = 2$)
such that **neighbors in high-D stay neighbors in low-D**.

### 2c. The Crowding Problem
When reducing dimensions (e.g., 2D to 1D), there is not enough "room" to faithfully place
all moderately-close neighbors.  In high-D, volume grows as $r^D$; in 2D, perimeter grows
only as $2\pi r$.  Points that were comfortably spread on a high-D shell get **crushed** together
in low-D -- unless we use a heavy-tailed kernel (the t-SNE solution).

### 2d. Stochastic
The embedding is found through a procedure that includes **randomness**:
- **Initialization:** $\{y_i\}$ start randomly (or from PCA)
- **Optimization:** iterative gradient descent with momentum
- **Probabilities:** neighborhoods are *soft* probabilities ($p_{j|i}$, $q_{ij}$), not hard assignments

> **What stays the same:** The objective $\text{KL}(P \| Q)$ is fixed.  Different runs may produce
> rotated/reflected solutions, but local structure should be consistent.

## 3. Problem Statement & Distance Concentration

**The dimensionality reduction problem:**
Find $y_i \in \mathbb{R}^d$ such that local structure is preserved and false Euclidean neighbors are separated.

### Why raw distances are tricky in high dimensions

As $D$ grows, all pairwise distances **concentrate** around the same value.
The "nearest" neighbor is barely closer than a random point -- Euclidean distance becomes meaningless.

**t-SNE's solution:** Abandon global distances. Force every point to have the same
*effective number of neighbors* (perplexity) by adapting a local bandwidth $\sigma_i$.

## 4. The Core t-SNE Pipeline

t-SNE proceeds in three steps:

| Step | Input to Output | Kernel |
|------|----------------|--------|
| **Step 1** | High-D distances to similarities $p_{ij}$ | Gaussian (per-point $\sigma_i$) |
| **Step 2** | Low-D distances to similarities $q_{ij}$ | Student-t (heavy tail) |
| **Step 3** | Minimize mismatch $\text{KL}(P \| Q)$ | Gradient descent on $\{y_i\}$ |

---

## 4a. Step 1 -- High-Dimensional Similarities

### Conditional Neighborhood Probabilities

For every point $x_i$, compute the probability that it picks $x_j$ as its neighbor:

$$p_{j|i} = \frac{\exp\left(-\|x_i - x_j\|^2 / 2\sigma_i^2\right)}{\sum_{k \neq i} \exp\left(-\|x_i - x_k\|^2 / 2\sigma_i^2\right)}, \quad p_{i|i} = 0$$

- **Numerator:** Gaussian similarity between $x_i$ and $x_j$
- **Denominator:** Normalizes so $\sum_j p_{j|i} = 1$
- **$\sigma_i$:** Adapts to local density (small in dense regions, large in sparse)

### Symmetrization

Because $\sigma_i \neq \sigma_j$, the probability is asymmetric: $p_{j|i} \neq p_{i|j}$.
We symmetrize:

$$\boxed{p_{ij} = \frac{p_{j|i} + p_{i|j}}{2n}} \quad \text{where } p_{ii} = 0 \text{ and } \sum_{i \neq j} p_{ij} = 1$$

This ensures outliers are not completely ignored -- if $x_j$ strongly claims $x_i$ as neighbor,
mutual attraction keeps them connected.

### Perplexity -- the user-facing "knob"

$$\text{Perp}(P_i) = 2^{H(P_i)} \quad \text{where } H(P_i) = -\sum_{j \neq i} p_{j|i} \log_2 p_{j|i}$$

t-SNE runs a **binary search** for each point to find $\sigma_i$ such that
the local entropy matches the target perplexity. Think of perplexity as a smooth $k$-NN.

````python
# -- 4b. Building P from scratch --
# We implement each step of Step 1 to see exactly how t-SNE works.

def p_conditional_for_point(X, i, sigma):
    """Compute p_{j|i} for a single point i with bandwidth sigma."""
    n = X.shape[0]
    dist2 = np.sum((X - X[i])**2, axis=1)
    dist2[i] = np.inf  # exclude self
    num = np.exp(-dist2 / (2.0 * sigma**2))
    denom = np.sum(num)
    if denom < 1e-30:
        return np.zeros(n)
    p = num / denom
    p[i] = 0.0
    return p

def perplexity_from_p(p):
    """Compute perplexity = 2^H(P) from a conditional distribution."""
    p = p[p > 0]
    H = -np.sum(p * np.log2(p))
    return 2**H

def sigma_for_target_perplexity(X, i, target_perp, tol=1e-3, max_iter=60):
    """Binary search on log(sigma) to achieve target perplexity for point i."""
    lo, hi = -10.0, 10.0
    for _ in range(max_iter):
        mid = 0.5 * (lo + hi)
        sigma = np.exp(mid)
        p = p_conditional_for_point(X, i, sigma)
        perp = perplexity_from_p(p)
        if abs(perp - target_perp) < tol:
            return sigma, perp
        if perp > target_perp:
            hi = mid
        else:
            lo = mid
    sigma = np.exp(0.5 * (lo + hi))
    return sigma, perplexity_from_p(p_conditional_for_point(X, i, sigma))

def compute_P_matrix(X, perplexity=30, tol=1e-3):
    """Compute the full symmetric P matrix with per-point bandwidth."""
    n = X.shape[0]
    Pcond = np.zeros((n, n), dtype=np.float64)
    sigmas = np.zeros(n, dtype=np.float64)
    for i in range(n):
        sig, _ = sigma_for_target_perplexity(X, i, target_perp=perplexity, tol=tol)
        sigmas[i] = sig
        Pcond[i] = p_conditional_for_point(X, i, sig)
    # Symmetrize: p_ij = (p_{j|i} + p_{i|j}) / 2n
    P = (Pcond + Pcond.T) / (2.0 * n)
    np.fill_diagonal(P, 0.0)
    P = np.maximum(P, 1e-12)
    P /= P.sum()
    return P, sigmas

def compute_Q_matrix(Y):
    """Student-t kernel: q_{ij} = (1 + ||y_i - y_j||^2)^{-1} / Z."""
    dist2 = pairwise_distances(Y, squared=True)
    np.fill_diagonal(dist2, np.inf)
    W = 1.0 / (1.0 + dist2)
    W[~np.isfinite(W)] = 0.0
    Q = W / np.sum(W)
    return np.maximum(Q, 1e-12)

print("P/Q matrix functions defined.")
print("  p_conditional_for_point(X, i, sigma)")
print("  sigma_for_target_perplexity(X, i, target_perp)")
print("  compute_P_matrix(X, perplexity)")
print("  compute_Q_matrix(Y)")
````

## 5. Step 2 -- Low-Dimensional Similarities ($q_{ij}$)

t-SNE uses a **Student t-distribution** (1 degree of freedom) in the low-D space:

$$q_{ij} = \frac{(1 + \|y_i - y_j\|^2)^{-1}}{\sum_{k \neq l}(1 + \|y_k - y_l\|^2)^{-1}}, \quad q_{ii} = 0$$

### Why change the kernel?

The Student-t has **heavy tails** -- it stays higher for much longer than the Gaussian.
This allows points to be placed *further apart* in low-D while maintaining the same
probability weight, solving the **crowding problem**.

## 6. Step 3 -- Minimize Mismatch: KL Divergence

$$\mathcal{L} = \text{KL}(P \| Q) = \sum_{i \neq j} p_{ij} \log \frac{p_{ij}}{q_{ij}}$$

### The Asymmetry of KL

| Scenario | $p_{ij}$ | $q_{ij}$ | Penalty |
|----------|---------|---------|---------|
| **Drop a true neighbor** | LARGE | small | **HUGE** |
| **Keep a false neighbor** | small | LARGE | **Near ZERO** |

> **Conclusion:** t-SNE is obsessed with keeping true neighbors together,
> but does not care much if false neighbors accidentally end up close.

### The Gradient -- "Springs and Magnets"

$$\frac{\partial \mathcal{L}}{\partial y_i} = 4 \sum_{j \neq i} (p_{ij} - q_{ij}) \cdot \frac{y_i - y_j}{1 + \|y_i - y_j\|^2}$$

- $p_{ij} > q_{ij}$: **attraction** (pull together) -- true neighbors not close enough
- $p_{ij} < q_{ij}$: **repulsion** (push apart) -- false neighbors too close

````python
# -- 6a. KL Divergence Asymmetry Demo --
# A toy 3-element example showing KL's one-sided penalty.

def KL(P, Q):
    P = np.clip(np.asarray(P, dtype=np.float64), 1e-12, 1.0)
    Q = np.clip(np.asarray(Q, dtype=np.float64), 1e-12, 1.0)
    return float(np.sum(P * np.log(P / Q)))

P1 = np.array([0.70, 0.20, 0.10])  # True distribution
Q_good              = np.array([0.65, 0.25, 0.10])  # Close to P
Q_bad_true_neighbor  = np.array([0.10, 0.45, 0.45])  # Breaks the big p_ij
Q_bad_false_neighbor = np.array([0.80, 0.10, 0.10])  # Overstates a small pair

kl_good = KL(P1, Q_good)
kl_bad_true = KL(P1, Q_bad_true_neighbor)
kl_bad_false = KL(P1, Q_bad_false_neighbor)

fig = go.Figure(go.Bar(
    x=["Q ~ P (good)", "Q drops true neighbor", "Q adds false neighbor"],
    y=[kl_good, kl_bad_true, kl_bad_false],
    marker_color=[COLORS["C"], COLORS["A"], COLORS["D"]],
    text=[f"{v:.4f}" for v in [kl_good, kl_bad_true, kl_bad_false]],
    textposition="outside"
))
fig.update_layout(title="KL Divergence Asymmetry: Dropping True Neighbors Is Expensive",
                  yaxis_title="KL(P || Q)", width=700, height=400,
                  template="plotly_white")
fig.show()

print(f"P = {P1}")
print(f"KL(P || Q_good)              = {kl_good:.4f}   <- Small penalty")
print(f"KL(P || Q_bad_true_neighbor) = {kl_bad_true:.4f}   <- HUGE penalty!")
print(f"KL(P || Q_bad_false_neighbor)= {kl_bad_false:.4f}   <- Small penalty")
````

````python
# -- 6b. Gradient Force Field Visualization --
# See the attraction/repulsion forces on a tiny system.

def compute_forces(P, Y):
    """Compute the t-SNE gradient force vectors."""
    Q = compute_Q_matrix(Y)
    dist2 = pairwise_distances(Y, squared=True)
    np.fill_diagonal(dist2, np.inf)
    w = 1.0 / (1.0 + dist2)
    w[~np.isfinite(w)] = 0.0

    F = np.zeros_like(Y)
    for i in range(Y.shape[0]):
        diff = Y[i] - Y                         # (n, 2)
        coeff = (P[i] - Q[i]) * w[i]            # (n,)
        F[i] = 4.0 * np.sum(coeff[:, None] * diff, axis=0)
    return F

# Tiny toy system
rng_toy = np.random.RandomState(1)
X_toy = StandardScaler().fit_transform(rng_toy.randn(12, 5))
P_toy, _ = compute_P_matrix(X_toy, perplexity=4)
Y_toy = rng_toy.randn(12, 2) * 0.01  # random init
F_toy = compute_forces(P_toy, Y_toy)

# Plot with Plotly (scatter + arrows)
fig = go.Figure()
fig.add_trace(go.Scatter(x=Y_toy[:,0], y=Y_toy[:,1], mode="markers+text",
                          marker=dict(size=10, color=COLORS["B"]),
                          text=[str(i) for i in range(12)],
                          textposition="top center", name="Points"))
# Draw arrows (negative gradient = direction of movement)
for i in range(len(Y_toy)):
    scale = 0.3
    fig.add_annotation(x=Y_toy[i,0] - F_toy[i,0]*scale,
                       y=Y_toy[i,1] - F_toy[i,1]*scale,
                       ax=Y_toy[i,0], ay=Y_toy[i,1],
                       xref="x", yref="y", axref="x", ayref="y",
                       showarrow=True, arrowhead=2, arrowsize=1.5,
                       arrowcolor=COLORS["A"])

fig.update_layout(title="Gradient Forces on a Toy System (arrows = negative grad direction)",
                  xaxis_title="y1", yaxis_title="y2",
                  width=650, height=500, template="plotly_white")
fig.show()
print("Arrows show how each point would move in one gradient step.")
print("Points with high p_ij (true neighbors) attract; others repel.")
````

## 7. Optimization in Practice

The gradient descent loop:

1. **Initialize** positions $\{y_i\}$ (random or PCA)
2. **Compute** all $q_{ij}$ from current positions
3. **Compute** gradient $\partial \mathcal{L} / \partial y_i$
4. **Update** $y_i \leftarrow y_i - \eta \nabla_{y_i}\mathcal{L} + \alpha \Delta y_i^{(t-1)}$
5. **Repeat** 250--1000 iterations

### Key Hyperparameters

| Parameter | Typical Range | Effect |
|-----------|--------------|--------|
| **Learning rate ($\eta$)** | 200--1000 (or `"auto"`) | Too small: blob; too large: explosion |
| **Early exaggeration** | x12 for first 250 iters | Forces tight clusters to form early |
| **Initialization** | `"pca"` recommended | PCA init: more stable, faster convergence |
| **Iterations** | 500--1000 | Too few: incomplete convergence |

````python
# -- 7a. Learning Rate Explorer --
# See how the learning rate affects convergence.

X_lr, color_lr, _ = make_dataset("blobs", n=400, seed=42)
X_lr = StandardScaler().fit_transform(X_lr)

rates = {"10 (too small)": 10,
         "200 (default)": 200,
         "500 (moderate)": 500,
         "5000 (too large)": 5000}

fig = make_subplots(rows=1, cols=4,
                    subplot_titles=list(rates.keys()))
for idx, (label, lr) in enumerate(rates.items()):
    Y = run_tsne(X_lr, perplexity=30, lr=lr, seed=42, n_iter=800)
    for c_val in np.unique(color_lr):
        mask = color_lr == c_val
        clr = PLOTLY_DISCRETE[int(c_val) % len(PLOTLY_DISCRETE)]
        fig.add_trace(go.Scatter(x=Y[mask,0], y=Y[mask,1], mode="markers",
                                  marker=dict(size=4, color=clr, opacity=0.7),
                                  showlegend=False),
                      row=1, col=idx+1)

fig.update_layout(title="Effect of Learning Rate on t-SNE", width=1100, height=350,
                  template="plotly_white")
for i in range(1, 5):
    fig.update_xaxes(showticklabels=False, row=1, col=i)
    fig.update_yaxes(showticklabels=False, row=1, col=i)
fig.show()
print("Too small -> blob (stuck in initial state).")
print("Too large -> scattered chaos (gradients overshoot).")
print("Moderate -> clear, separated clusters.")
````

## 9. Common Pitfalls -- What You Can and Cannot Trust

> **The Golden Rule:** t-SNE is a **topology map** (who is next to whom),
> **not a metric map**. Global distances, axes, and cluster sizes are all distorted.

### Pitfall 1: Axes Are Meaningless
The KL objective is invariant to rotation, translation, and scale.
**Never say** "Feature X increases as you move right."

### Pitfall 2: Global Distances Are Illusions
The empty space between clusters is arbitrary.
You **cannot** conclude that cluster A is closer to B than to C.

### Pitfall 3: Sizes and Densities Are Distorted
The local bandwidth $\sigma_i$ expands sparse regions and shrinks dense ones.
All clusters end up looking roughly the same size.

### Pitfall 4: Artifacts (Tearing Continuous Data)
Very low perplexity can fragment a perfectly continuous distribution into false "islands."

### Pitfall 5: Cherry-Picking (Confirmation Bias)
Because t-SNE has tunable hyperparameters and is stochastic, it is easy to re-run
until you see the pattern you *want* to see.
**Fix:** Sweep parameters; only trust structures stable across runs.

````python
# -- 9a. Pitfall 1 & 5: Axes are meaningless + Different seeds --
# Same data, different random seeds -> rotated/reflected but same local structure.

X_pit, color_pit, _ = make_dataset("blobs", n=400, seed=42)
X_pit = StandardScaler().fit_transform(X_pit)

fig = make_subplots(rows=1, cols=4, subplot_titles=[f"Seed {s}" for s in [0, 1, 2, 3]])
for idx, seed in enumerate([0, 1, 2, 3]):
    Y = run_tsne(X_pit, perplexity=30, seed=seed)
    for c_val in np.unique(color_pit):
        mask = color_pit == c_val
        clr = PLOTLY_DISCRETE[int(c_val) % len(PLOTLY_DISCRETE)]
        fig.add_trace(go.Scatter(x=Y[mask,0], y=Y[mask,1], mode="markers",
                                  marker=dict(size=4, color=clr, opacity=0.7),
                                  showlegend=(idx==0), name=f"C{int(c_val)}"),
                      row=1, col=idx+1)

fig.update_layout(title="Different Seeds -> Different Orientations (Same Local Structure)",
                  width=1100, height=350, template="plotly_white")
for i in range(1, 5):
    fig.update_xaxes(showticklabels=False, row=1, col=i)
    fig.update_yaxes(showticklabels=False, row=1, col=i)
fig.show()
print("Each run produces a different rotation/reflection.")
print("The AXES have no intrinsic meaning -- only within-cluster structure matters.")
````

````python
# -- 9b. Pitfall 2: Global Distances Are Illusions --
# Cluster centroids in the original vs embedded space.

X_gd, color_gd, _ = make_dataset("blobs", n=400, seed=42)
X_gd = StandardScaler().fit_transform(X_gd)
Y_gd = run_tsne(X_gd, perplexity=30, seed=42)

labels_unique = np.unique(color_gd)
centroids_orig = np.array([X_gd[color_gd == c].mean(axis=0) for c in labels_unique])
centroids_emb  = np.array([Y_gd[color_gd == c].mean(axis=0) for c in labels_unique])

d_orig = squareform(pdist(centroids_orig))
d_emb  = squareform(pdist(centroids_emb))

# Normalize for comparison
d_orig_n = d_orig / d_orig.max()
d_emb_n  = d_emb / d_emb.max()

cluster_names = [f"C{int(c)}" for c in labels_unique]
fig = make_subplots(rows=1, cols=2,
                    subplot_titles=["Original Inter-Cluster Distances (normalized)",
                                    "t-SNE Inter-Cluster Distances (normalized)"])
fig.add_trace(go.Heatmap(z=d_orig_n, x=cluster_names, y=cluster_names,
                          colorscale="Blues", showscale=False,
                          text=np.round(d_orig_n, 2), texttemplate="%{text}"),
              row=1, col=1)
fig.add_trace(go.Heatmap(z=d_emb_n, x=cluster_names, y=cluster_names,
                          colorscale="Reds", showscale=False,
                          text=np.round(d_emb_n, 2), texttemplate="%{text}"),
              row=1, col=2)
fig.update_layout(title="Pitfall 2: Inter-Cluster Distances Are NOT Preserved",
                  width=900, height=380, template="plotly_white")
fig.show()
print("The relative ordering and magnitudes of inter-cluster distances change!")
````

## 10. Validating Your Embedding

Never trust a t-SNE plot blindly. Use these sanity checks:

| Check | How | What it tells you |
|-------|-----|-------------------|
| **kNN Overlap (Jaccard)** | Compare k-NN in high-D vs low-D | Local structure preservation |
| **Trustworthiness** | `sklearn.manifold.trustworthiness` | How well low-D respects high-D neighbors |
| **Stability** | Multiple seeds, same perplexity | Whether clusters are real or artifacts |
| **PCA Baseline** | Run PCA first | If PCA already separates, t-SNE is unnecessary |

````python
# -- 10a. kNN Jaccard Overlap --
# For each point, compute the Jaccard similarity between its k-NN
# in the original space and in the embedding.

def knn_jaccard_overlap(X_high, Y_low, k=15):
    """Compute per-point Jaccard overlap of k-NN sets."""
    nn_high = NearestNeighbors(n_neighbors=k).fit(X_high)
    nn_low  = NearestNeighbors(n_neighbors=k).fit(Y_low)
    idx_high = nn_high.kneighbors(return_distance=False)
    idx_low  = nn_low.kneighbors(return_distance=False)

    jaccard = np.zeros(len(X_high))
    for i in range(len(X_high)):
        s_h = set(idx_high[i])
        s_l = set(idx_low[i])
        jaccard[i] = len(s_h & s_l) / len(s_h | s_l)
    return jaccard

X_val, color_val, _ = make_dataset("blobs", n=400, seed=42)
X_val = StandardScaler().fit_transform(X_val)
Y_val = run_tsne(X_val, perplexity=30, seed=42)

def demo_knn_overlap(k=15):
    jacc = knn_jaccard_overlap(X_val, Y_val, k=k)
    fig = make_subplots(rows=1, cols=2,
                        subplot_titles=[f"Embedding colored by Jaccard (k={k})",
                                        f"Jaccard Distribution (mean={jacc.mean():.3f})"])
    fig.add_trace(go.Scatter(x=Y_val[:,0], y=Y_val[:,1], mode="markers",
                              marker=dict(size=5, color=jacc, colorscale="RdYlGn",
                                          colorbar=dict(title="Jaccard"),
                                          opacity=0.8)),
                  row=1, col=1)
    fig.add_trace(go.Histogram(x=jacc, nbinsx=30, marker_color=COLORS["B"]),
                  row=1, col=2)
    fig.update_layout(title=f"kNN Overlap Validation (k={k})",
                      width=950, height=400, template="plotly_white")
    fig.show()

interact(demo_knn_overlap,
         k=widgets.IntSlider(min=5, max=50, value=15, description="k:"))
````

````python
# -- 10b. Trustworthiness Sweep --
# sklearn.manifold.trustworthiness measures how well low-D respects high-D neighbors.

X_trust = PCA(n_components=40).fit_transform(X_digits)

perps_tw = [5, 10, 15, 20, 30, 40, 50]
scores_tw, times_tw = [], []

print("Computing trustworthiness across perplexities...")
for p in perps_tw:
    t0 = perf_counter()
    Y_tw = run_tsne(X_trust, perplexity=p, seed=0, lr=200, n_iter=800)
    dt = perf_counter() - t0
    tw = trustworthiness(X_trust, Y_tw, n_neighbors=10)
    scores_tw.append(tw)
    times_tw.append(dt)
    print(f"  Perp={p:>2}: trustworthiness@10 = {tw:.4f}, time = {dt:.1f}s")

fig = make_subplots(rows=1, cols=2,
                    subplot_titles=["Trustworthiness@10 vs Perplexity",
                                    "Runtime vs Perplexity"])
fig.add_trace(go.Scatter(x=perps_tw, y=scores_tw, mode="lines+markers",
                          marker=dict(color=COLORS["B"])), row=1, col=1)
fig.add_trace(go.Bar(x=[str(p) for p in perps_tw], y=times_tw,
                      marker_color=COLORS["D"]), row=1, col=2)
fig.update_layout(title="Embedding Quality: Trustworthiness vs Perplexity",
                  width=950, height=400, template="plotly_white",
                  showlegend=False)
fig.update_yaxes(title_text="Trustworthiness", row=1, col=1)
fig.update_yaxes(title_text="Seconds", row=1, col=2)
fig.show()
````

## 11. Method Comparison: PCA / MDS / t-SNE / UMAP

| Method | Type | Preserves | Speed |
|--------|------|-----------|-------|
| **PCA** | Linear | Global variance | Fast |
| **MDS** | Nonlinear | Global distances | Moderate |
| **t-SNE** | Nonlinear | Local neighborhoods | Moderate |
| **UMAP** | Nonlinear | Local + some global | Fast |

> **Best practice:** Run both t-SNE and UMAP.
> If both find the same clusters, confidence is high.

````python
# -- 11a. Method Comparison Dashboard --

def demo_comparison(dataset="swiss_roll", n_points=400):
    X, color, title = make_dataset(dataset, n=n_points, seed=42)
    X_s = StandardScaler().fit_transform(X)
    is_cont = not np.array_equal(color, color.astype(int))

    methods = {}

    # PCA
    t0 = perf_counter()
    methods["PCA"] = PCA(n_components=2).fit_transform(X_s)
    methods["PCA_t"] = perf_counter() - t0

    # MDS
    t0 = perf_counter()
    methods["MDS"] = MDS(n_components=2, random_state=42,
                         normalized_stress="auto").fit_transform(X_s)
    methods["MDS_t"] = perf_counter() - t0

    # t-SNE
    t0 = perf_counter()
    methods["t-SNE"] = run_tsne(X_s, perplexity=30, seed=42)
    methods["t-SNE_t"] = perf_counter() - t0

    # UMAP
    try:
        import umap
        t0 = perf_counter()
        methods["UMAP"] = umap.UMAP(n_neighbors=15, min_dist=0.1,
                                     random_state=42).fit_transform(X_s)
        methods["UMAP_t"] = perf_counter() - t0
    except ImportError:
        methods["UMAP"] = None

    names = ["PCA", "MDS", "t-SNE", "UMAP"]
    fig = make_subplots(rows=1, cols=4,
                        subplot_titles=[f"{n} ({methods.get(n+'_t', 0):.1f}s)"
                                        if methods.get(n) is not None else f"{n} (N/A)"
                                        for n in names])

    for idx, name in enumerate(names):
        Y = methods.get(name)
        if Y is None:
            fig.add_annotation(text="Not installed", x=0.5, y=0.5,
                               xref=f"x{idx+1} domain", yref=f"y{idx+1} domain",
                               showarrow=False, font=dict(size=14))
            continue
        if is_cont:
            fig.add_trace(go.Scatter(
                x=Y[:,0], y=Y[:,1], mode="markers",
                marker=dict(size=4, color=color, colorscale="Spectral", opacity=0.7),
                showlegend=False), row=1, col=idx+1)
        else:
            for c_val in np.unique(color):
                mask = color == c_val
                clr = PLOTLY_DISCRETE[int(c_val) % len(PLOTLY_DISCRETE)]
                fig.add_trace(go.Scatter(
                    x=Y[mask,0], y=Y[mask,1], mode="markers",
                    marker=dict(size=4, color=clr, opacity=0.7),
                    showlegend=False), row=1, col=idx+1)

    fig.update_layout(title_text=f"Method Comparison: {title}", width=1200, height=380,
                      template="plotly_white")
    for i in range(1, 5):
        fig.update_xaxes(showticklabels=False, row=1, col=i)
        fig.update_yaxes(showticklabels=False, row=1, col=i)
    fig.show()

interact(demo_comparison,
         dataset=widgets.Dropdown(
             options=["swiss_roll", "s_curve", "blobs", "moons", "concentric"],
             value="swiss_roll", description="Dataset:"),
         n_points=widgets.IntSlider(min=200, max=600, step=100, value=400,
                                    description="N:"))
````

## 12. Real-World Application: MNIST Handwritten Digits

The complete t-SNE pipeline on real data:

1. **Load** the digits dataset (8x8 pixel images, 64 features, 1797 samples)
2. **Scale** with StandardScaler
3. **Reduce** with PCA to 30 components
4. **Embed** with t-SNE (perplexity=30, PCA init)
5. **Visualize** with interactive Plotly scatter

````python
# -- 12a. MNIST t-SNE Pipeline --

print("Step 1: Loading digits dataset...")
digits = load_digits()
print(f"  Shape: {digits.data.shape} ({digits.data.shape[0]} samples, {digits.data.shape[1]} features)")

print("Step 2: Standardizing features...")
X_mnist = StandardScaler().fit_transform(digits.data)

print("Step 3: PCA to 30 dimensions...")
X_mnist_pca = PCA(n_components=30).fit_transform(X_mnist)
var_explained = PCA(n_components=30).fit(X_mnist).explained_variance_ratio_.sum()
print(f"  Variance retained: {var_explained:.1%}")

print("Step 4: Running t-SNE (perplexity=30, 1000 iterations)...")
t0 = perf_counter()
Y_mnist = TSNE(n_components=2, perplexity=30, learning_rate=200,
               max_iter=1000, init="pca", random_state=42).fit_transform(X_mnist_pca)
dt_mnist = perf_counter() - t0
print(f"  Done in {dt_mnist:.1f}s")

tw_mnist = trustworthiness(X_mnist_pca, Y_mnist, n_neighbors=10)
print(f"  Trustworthiness@10: {tw_mnist:.4f}")

print("Step 5: Interactive visualization...")
fig = go.Figure()
for d in range(10):
    mask = digits.target == d
    fig.add_trace(go.Scatter(
        x=Y_mnist[mask, 0], y=Y_mnist[mask, 1], mode="markers",
        name=f"Digit {d}",
        marker=dict(size=5, color=PLOTLY_DISCRETE[d], opacity=0.7),
        text=[f"Digit {d}"] * mask.sum(), hoverinfo="text"))

fig.update_layout(
    title=f"t-SNE of Handwritten Digits (hover to explore!) -- Trust@10={tw_mnist:.3f}",
    xaxis_title="t-SNE 1", yaxis_title="t-SNE 2",
    width=850, height=650, template="plotly_white",
    legend=dict(title="Digit"))
fig.show()

print("\nClick legend entries to toggle digits on/off.")
print("Zoom into boundary regions (e.g., 4 vs 9, 3 vs 8) to explore confusion zones.")
````

Full parameter exploration -- try different settings and see how they affect the embedding.
The dashboard computes t-SNE in real-time, so each run may take a few seconds.

````python
# -- 13a. Full Parameter Dashboard --

# Prepare data for the dashboard
X_dash = PCA(n_components=40).fit_transform(
    StandardScaler().fit_transform(digits.data))
y_dash = digits.target

def tsne_dashboard(perplexity=30, learning_rate=200, init="pca",
                   n_iter=800, seed=42, early_exag=12.0):
    t0 = perf_counter()
    Y = TSNE(n_components=2,
             perplexity=perplexity,
             learning_rate=learning_rate if learning_rate != 0 else "auto",
             max_iter=n_iter,
             init=init,
             random_state=seed,
             early_exaggeration=early_exag
    ).fit_transform(X_dash)
    dt = perf_counter() - t0

    tw = trustworthiness(X_dash, Y, n_neighbors=10)

    fig = go.Figure()
    for d in range(10):
        mask = y_dash == d
        fig.add_trace(go.Scatter(
            x=Y[mask,0], y=Y[mask,1], mode="markers",
            name=f"{d}", marker=dict(size=4, color=PLOTLY_DISCRETE[d], opacity=0.7)))

    fig.update_layout(
        title=f"perp={perplexity}, lr={learning_rate}, init={init}, "
              f"iter={n_iter}, seed={seed}, exag={early_exag:.0f} -- "
              f"time={dt:.1f}s, trust@10={tw:.3f}",
        width=800, height=550, template="plotly_white",
        legend=dict(title="Digit", orientation="h", y=-0.1))
    fig.show()

interact(tsne_dashboard,
         perplexity=widgets.IntSlider(min=5, max=100, value=30, description="Perplexity:"),
         learning_rate=widgets.IntSlider(min=0, max=2000, step=50, value=200,
                                         description="LR (0=auto):"),
         init=widgets.Dropdown(options=["pca", "random"], value="pca", description="Init:"),
         n_iter=widgets.IntSlider(min=250, max=2000, step=50, value=800,
                                   description="Iterations:"),
         seed=widgets.IntSlider(min=0, max=100, value=42, description="Seed:"),
         early_exag=widgets.FloatSlider(min=1, max=30, step=1, value=12,
                                         description="Exaggeration:"))
````

## 14. Summary: The 5 Rules of Thumb for t-SNE

| # | Rule | Details |
|---|------|---------|
| **1** | **Pre-process** | Scale your features to standard variance |
| **2** | **Denoise** | Run PCA first (reduce to 30--50 dims) |
| **3** | **Sweep** | Test multiple perplexity values (5 to 50) |
| **4** | **Verify** | Run multiple random seeds to check stability |
| **5** | **Interpret Locally** | Ignore global distances, cluster sizes, and axes |

### Reproducibility Checklist

When publishing a t-SNE plot, you **must** report:
- Scaling method
- Number of PCA dimensions
- Perplexity value
- Learning rate
- Number of iterations
- Initialization method
- Random seed
- Software and version

### The Golden Takeaway

> t-SNE is a visualization tool for **hypothesis generation**, not mathematical proof.
> Use it to **Explore**, **Hypothesize**, then **Validate** with rigorous methods.

---

### The Standard Operating Procedure

```
1. Scale -> StandardScaler
2. Reduce -> PCA to ~50 dims
3. Initialize -> PCA init for t-SNE
4. Sweep -> Perplexity {5, 10, 30, 50}
5. Verify -> 3+ random seeds per setting
6. Conclude -> Only trust patterns stable across all runs
```
